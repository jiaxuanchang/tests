```python
import numpy as np
import xarray as xr
import pandas as pd
```

```python
import datetime
```

```python
da = xr.DataArray(
    np.random.randint(1000, size=1827),
    coords=[pd.date_range("2000-01-01", "2004-12-31", freq="D")],
    dims="time",)
```

```python
da
```

PeakEeveyYear is the maximum value in that year.

```python
PeakEveryYear = da.groupby("time.year").max()
```

```python
PeakEveryYear
```

Here we test how to find the time with peak value in the group of 2000 year

```python
(da.groupby("time.year")[2000].where(da.groupby("time.year")[2000]==PeakEveryYear[0], drop=True)).time
```

So we see the date is 2000-12-30 with the peak value of 999. Then we can get the day of the year by 'time.dt.dayofyear'.

```python
# ind1 is the day of the year of 2000. 
ind1=(da.groupby("time.year")[2000].where(da.groupby("time.year")[2000]==PeakEveryYear[0], drop=True)).time.dt.dayofyear.values
print(ind1)
```

```python
len(da.groupby('time.year')) #how many groups
```

```python
ii=0
ind=np.zeros(len(da.groupby('time.year')))
for label, group in da.groupby('time.year'):
    print(label)
    ind[ii]=da.groupby("time.year")[label].where(da.groupby("time.year")[label]==PeakEveryYear[ii], drop=True).time[0].dt.dayofyear.values
    ii=ii+1
    
print(ind)
```
